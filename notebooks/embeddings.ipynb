```python
import pandas as pd
import numpy as np
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")                       # Free 384 dimensional embeddings model
```

```python
df = pd.read_csv("../data/raw/pubmed_papers.csv")
```

```python
sample_text = df.iloc[0]['Abstract']
sample_text[:300]
```

```python
embedding = model.encode(sample_text)
```

```python
print(type(embedding))
embedding.shape
```

```python
print(embedding[:10])           #Texts into vectors
```

For understanding encoding 

```python
text1 = "lung cancer biomarker"

text2 = "biomarkers for lung cancer diagnosis"

text3 = "how to cook pasta"
```

```python
emb1 = model.encode(text1)
emb2 = model.encode(text2)
emb3 = model.encode(text3)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

print(cosine_similarity([emb1],[emb2]))
print(cosine_similarity([emb1],[emb3]))
```
