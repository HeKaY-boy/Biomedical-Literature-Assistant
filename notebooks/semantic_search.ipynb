```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(("../data/raw/pubmed_papers.csv"))
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
embedding = model.encode(df['Abstract'].to_list(), show_progress_bar=True)
```

```python
df.shape
```

```python
embedding.shape
```

```python
query = 'lung cancer diagnosis biomarkers'

embedding_query = model.encode(query)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity([embedding_query], embedding)

similarities.shape
```

```python
best_sim = similarities.argmax()
title = df.iloc[best_sim]['Title']
print(best_sim)
title
```

```python
top_5_best_indx = np.argsort(similarities[0])[-5:][::-1]
top_5_best_indx
```

```python
for rank, id in enumerate(top_5_best_indx, start=1):
    title = df.iloc[id]['Title']
    print(f"{rank}.{title}")
    
    sim = similarities[0][id]
    print(f"Similarity: {sim:.4f}\n")
    rank +=1
    
```

```python
def search_paper(query, top_k =5):
    embedding_query = model.encode(query)
    similarities = cosine_similarity([embedding_query], embedding)
    results= {'PMID': [], 'Title':[], 'Abstract':[]}
    for id in top_k:
        pmid = df.iloc[id]['PMID']
        title = df.iloc[id]['Title']
        abstract = df.iloc[id]['Abstract']

        results['PMID'].append(pmid)
        results['Title'].append(title)
        results['Abstract'].append(abstract)
    return results

    
```
